6 - Faça uma análise dos dados faltantes e trate-os
<br>Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
display(dataset2.isnull().sum())#checando quais colunas possuem dados faltantes e a quantidade
```

```python
#fazendo uma cópia do DataFrame antes de incluir valores nos dados faltantes caso seja necessária futuramente
dataset_dadosfaltantes = dataset2.copy(deep=True)
```

```python
#Para os ativos, preenchi com '0' visto que possivelmente o a falta do valor em determinado ativo signifique que investidor não tenha aplicações neste ativo:
dataset2[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06']] = dataset2[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06']] .fillna(0)
display(dataset2.head(6))#checando se a substituição ocorreu como esperado
```

```python
#Ambas as notas parecem ter uma distribuição normal, por isso escolhi a média da coluna como forma de substituir.
display(dataset2['Nota1'].describe())#distirbuição normal

display(dataset2['Nota2']. describe())#distribuição normal

media_nota1 = dataset2['Nota1'].mean()
dataset2['Nota1'] = dataset2['Nota1'].fillna(media_nota1)#preenchendo com a média da coluna Nota1

media_nota2 = dataset2['Nota2'].mean()
dataset2['Nota2'] = dataset2['Nota2'].fillna(media_nota2)#preenchendo com a média da coluna Nota2
```

```python
display(dataset2.head())#Checando os dados
```

7 - Manipule os dados a seguir:
 - Criar coluna "Total" somando o valor de todos os ativos por investidor
 - Criar coluna "Media" com a média das notas 1 e 2
 - Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
 - Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
 - Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
 - Transformar os dados da coluna Media em inteiro
 - Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50

```python
# Total de valores dos ativos por investidor
dataset2['Soma_ativos'] = dataset2['Ativo01'] + dataset2['Ativo02'] + dataset2['Ativo03'] + dataset2['Ativo04'] + dataset2['Ativo05'] + dataset2['Ativo06'] + dataset2['Ativo07']
```

```python
#Média das notas
dataset2['Media'] = (dataset2['Nota1'] + dataset2['Nota2'])/2
```

```python
#Coluna "Dominio"
dataset2[['Nomeemail', 'Dominio']] = dataset2['Email'].str.split('@', expand=True)
dataset2 = dataset2.drop(columns = 'Nomeemail')
```

```python
#Mudando o nome da coluna "Salário" para "Renda" e tratando a coluna(sem $ e numérica)
dataset2 = dataset2.rename(columns={
    'Salario': 'Renda'
})
dataset2['Renda'] = dataset2['Renda'].str.replace('$', '').astype(float)
dataset2['Renda'].dtypes #checando
```

```python
#Mudar os dados da coluna sexo para Masculino e Feminino (usar o map)
dataset2['Sexo'] = dataset2['Sexo'].map({'Fem': 'Feminino', 'Masc': 'Masculino'})
```

```python
#Transformar os dados da coluna 'Media' em inteiro
dataset2['Media'] = dataset2['Media'].astype(int)
```

```python
#Criar coluna NPS
dataset2['NPS'] = dataset2.Media.apply(lambda x: 'promotor' if x >= 80 else ('neutro' if x> 50 and x<80 else 'detrator'))
```
